### Quem são os 10 jogadores com mais partidas nos últimos 10 anos?

```python
# Filtrando os dados para o período de 2012 a 2022
games_filtered = Games[(Games['SEASON'] >= 2012) & (Games['SEASON'] <= 2022)]

# Merge das bases de dados Games e Games_details para obter informações detalhadas dos jogos
merged_data = pd.merge(Games_details, games_filtered, how='inner', on='GAME_ID')

# Agrupamento por jogador e contagem de jogos
player_games_count = merged_data.groupby('PLAYER_NAME')['GAME_ID'].nunique().reset_index()
player_games_count.columns = ['PLAYER_NAME', 'TOTAL_GAMES']

# Ordenando os jogadores pela contagem de jogos em ordem decrescente
sorted_players = player_games_count.sort_values(by='TOTAL_GAMES', ascending=False)

# Imprimindo os jogadores com mais jogos nos últimos 10 anos
print(sorted_players.head(10))
```

```python
# Filtrando os 10 jogadores com mais jogos
top_10_players = sorted_players.head(10)

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_players['PLAYER_NAME'], top_10_players['TOTAL_GAMES'], color='red')
plt.xlabel('Jogador')
plt.ylabel('Número de Jogos')
plt.title('Top 10 Jogadores com Mais Jogos nos Últimos 10 Anos')
plt.xticks(rotation=45, ha='right')  # Rotaciona os nomes dos jogadores para melhorar a legibilidade

# Adicionando rótulos de dados em cada barra
for bar in bars:
    yval = bar.get_height()
    plt.annotate(round(yval, 2),
                 xy=(bar.get_x() + bar.get_width() / 2, yval),
                 xytext=(0, 3),  # Deslocamento vertical para melhor posicionamento
                 textcoords="offset points",
                 ha='center', va='bottom')

# Exibindo o gráfico
plt.show()
```

### Quais times apresentaram os melhores desempenhos no período abrangido pelos dados em nossa base?

```python
# Selecione as linhas onde "G" é igual a 82 na base de dados Ranking
# Esse passo é necessário para identificarmos somente os valores finais das equipes em cada temporada, no entanto ira excluir as temporadas 2011-2012 e 2020-2021 que tiveram menos de 82 jogos por diversas circunstâncias
filtered_ranking = Ranking[Ranking["G"] == 82]

# Faça a junção das bases de dados filtered_ranking e Teams com base na coluna "TEAM_ID"
merged_data = filtered_ranking.merge(Teams, on="TEAM_ID")

# Selecionando apenas as colunas necessárias
colunas_desejadas = ["TEAM_ID", "TEAM", "SEASON_ID", "NICKNAME", "W_PCT"]
merged_data_filtered = merged_data[colunas_desejadas]

# Excluindo os valores duplicados com base em todas as colunas
merged_data_final = merged_data_filtered.drop_duplicates()

# Exiba as últimas 10 linhas da base de dados
print(merged_data_final.head(10))
```

```python
# Criando uma coluna para armazenar a média W_PCT de cada time
merged_data_final['W_PCT_MEAN'] = merged_data_final.groupby('TEAM_ID')['W_PCT'].transform('mean')

# Mantenha apenas as colunas necessárias
result = merged_data_final[['TEAM', 'NICKNAME', 'W_PCT_MEAN']]

# Remova as linhas duplicadas, considerando apenas a primeira ocorrência de cada equipe
result = result.drop_duplicates(subset='NICKNAME')

# Ordene a base de dados pelo W_PCT_MEAN em ordem crescente
result_sorted = result.sort_values(by='W_PCT_MEAN')

# Exiba o resultado
result_sorted.head(10)
```

```python
# Calcule a média geral de W_PCT_MEAN
media_geral = result_sorted['W_PCT_MEAN'].mean()

# Crie um gráfico de barras horizontais
plt.figure(figsize=(10, 8))
plt.barh(result_sorted['NICKNAME'], result_sorted['W_PCT_MEAN'], color='green', label='Média por Equipe')
plt.axvline(x=media_geral, color='red', linestyle='--', label='Média Geral')

# Adicione rótulos de dados
for index, value in enumerate(result_sorted['W_PCT_MEAN']):
    plt.text(value, index, f'{value:.3f}', va='center')

plt.xlabel('Percentual de vitórias')
plt.title('Média percentual de vitórias de cada equipe')
plt.legend()
plt.grid(axis='x')

# Exiba o gráfico
plt.show()
```

### Como se comparam as médias de desempenho dos MVPs nos últimos 10 anos em relação aos demais jogadores?

Para abordar essa questão, será necessário incorporar uma nova base de dados contendo informações sobre os jogadores que foram eleitos MVP em suas respectivas temporadas.

Essa base de dados possui as colunas:

Player: Nome do jogador eleito MVP
Season: Temporada na qual o jogador foi eleito MVP

```python
# Importação da nova base de dados

Mvp = pd.read_csv('/content/Mvps.csv')
```

```python
# Merge das tabelas usando a coluna GAME_ID como chave
Games_details_s = pd.merge(Games_details, Games[['GAME_ID', 'SEASON']], on='GAME_ID', how='left')

# Junte os datasets para identificar as temporadas de MVP
merged_data = Games_details_s.merge(Mvp, how="left", left_on=["PLAYER_NAME", "SEASON"], right_on=["Player", "Season"])

# Filtrando os dados para o período de 2012 a 2022
merged_data = merged_data[(merged_data['SEASON'] >= 2012) & (merged_data['SEASON'] <= 2022)]

# Selecionando apenas as estatísticas desejadas
selected_stats = ["PTS", "REB", "AST"]
merged_data = merged_data[['Player'] + selected_stats]

# Calcule as médias para jogadores MVP
mvp_averages = merged_data[merged_data["Player"].notnull()].groupby("Player").mean()[selected_stats]

# Calcule as médias para outros jogadores
other_players_averages = merged_data[merged_data["Player"].isnull()].mean()[selected_stats]

# Médias dos jogadores MVP
mvp_means = mvp_averages.mean().values

# Médias dos outros jogadores
other_players_means = other_players_averages.values

# Nomes das estatísticas
stat_labels = selected_stats

# Número de estatísticas
num_stats = len(stat_labels)

# Criação de um array com os índices das barras
bar_width = 0.35
index = np.arange(num_stats)

# Criar barras para jogadores MVP
plt.bar(index, mvp_means, bar_width, label='Jogadores MVP', color='b')

# Adicionar rótulos de dados nas barras para jogadores MVP
for i, value in enumerate(mvp_means):
    plt.text(i, value + 0.1, f'{value:.2f}', ha='center', va='bottom')

# Criar barras para outros jogadores
plt.bar(index + bar_width, other_players_means, bar_width, label='Outros Jogadores', color='r')

# Adicionar rótulos de dados nas barras para outros jogadores
for i, value in enumerate(other_players_means):
    plt.text(i + bar_width, value + 0.1, f'{value:.2f}', ha='center', va='bottom')

# Adicionar rótulos ao gráfico
plt.xlabel('Estatísticas')
plt.ylabel('Valor Médio')
plt.title('Comparação das Médias de Jogadores MVP e Outros Jogadores (2012-2022)')
plt.xticks(index + bar_width / 2, stat_labels)

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
